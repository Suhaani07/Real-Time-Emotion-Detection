```python
#importing packages
from keras.utils import to_categorical
from keras_preprocessing.image import load_img
from keras.models import Sequential
from keras.layers import Dense, Conv2D, Dropout, Flatten, MaxPooling2D
import os
import pandas as pd
import numpy as np
```

```python
#path to store train and test datasets
TRAIN_DIR='images/train'
TEST_DIR='images/test'
```

```python
#creating path-label dataframe
def createdataframe(dir):
    image_paths=[]
    labels=[]
    #iterating over sub folders like angry,sad
    for label in os.listdir(dir):
        #iterating over images in the subfolders like 0.png, 1.png
        #joining dir name and klabel to form the sub folder name :
        #Eg : dir=images/train label = angry. Thus iterating over images inside images/train/angry as path
        for imagename in os.listdir(os.path.join(dir,label)):
            # Eg : adding path as images/train/angry/1.png
            image_paths.append(os.path.join(dir,label,imagename))
            # Eg : adding label as angry
            labels.append(label)
        print(label,"completed")
    return image_paths, labels
        
```

```python
#making training dataframe
train=pd.DataFrame()
train['image'],train['label']=createdataframe(TRAIN_DIR)
```

```python
#checking dataframe
print(train)
```

```python
#making testing dataframe
test=pd.DataFrame()
test['image'],test['label']=createdataframe(TEST_DIR)
```

```python
#checking dataframe
print(test)
```

```python
#importing tqdm
from tqdm.notebook import tqdm
```

```python
#processing images
#tqdm is used to give visual progress bar to show how much data has been processed
def extract_features(images):
    features=[]
    for image in tqdm(images):
        #loading image as grayscale
        img=load_img(image, color_mode = "grayscale")
        #converting image into 2d array
        img=np.array(img)
        features.append(img)
    #converting the whole list into array
    features=np.array(features)
    #converting into 1d array where 48*48 is size of image as 1 as 2d image
    features=features.reshape(len(features),48,48,1)
    return features
```

```python
#processing images in train dataframe
train_feature=extract_features(train['image'])
```

```python
#processing images in test dataframe
test_feature=extract_features(test['image'])
```

```python
#normalizing the images, i.e. converting pixel values in range of 0-1
x_train=train_feature/255.0
x_test=test_feature/255.0
print(x_train)
print(x_test)
```

```python
#Supervised learning : giving label with input thus importing LabelEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
#LabelEncoder converts Labels into numerical label. Eg : angry -> 0, sad-> 1 
le=LabelEncoder()
le.fit(train['label'])
```

```python
y_train=le.transform(train['label'])
y_test=le.transform(test['label'])
print(y_train)
print(y_test)
```

```python
#converting to binary matrix, we have 7 different labels
y_train=to_categorical(y_train,num_classes=7)
y_test=to_categorical(y_test,num_classes=7)
print(y_train)
print(y_test)
```

```python
#making model as x and y are ready

#sequential architecture for a classification task
#In Keras, the sequential model is a linear stack of layers, where you can add one layer at a time
model=Sequential()

#convolutional layer

#Conv2D() creates a 2D convolutional layer with 128 filters and a filter size of 3x3
#These filters are used to scan over the input image to extract features
#activation='relu' specifies that the Rectified Linear Unit (ReLU) activation function will be applied after the convolution operation
#input_shape=(48,48,1) defines the input shape of the images the model will receive. It's a 48x48 pixel grayscale image (1 channel)
model.add(Conv2D(128,kernel_size=(3,3),activation='relu', input_shape=(48,48,1)))

#Max-pooling is used to reduce the dimensions of the feature maps produced by the previous convolutional layer
model.add(MaxPooling2D(pool_size=(2,2)))

#Dropout prevent overfitting by randomly deleting a fraction of data
model.add(Dropout(0.4))

# add more convolutional, maxpooling and dropout layers
model.add(Conv2D(256,kernel_size=(3,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.4))

model.add(Conv2D(512,kernel_size=(3,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.4))

model.add(Conv2D(512,kernel_size=(3,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.4))

#transform the 2D feature maps(image data) into a 1D array
model.add(Flatten())

#fully connected layers
model.add(Dense(512,activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(256,activation='relu'))
model.add(Dropout(0.3))

#output layer
#The final dense layer has 7 labels 
#The 'softmax' activation function is used to convert the output into label probabilities. Eg : 0.9 fo angry, 0.1 for sad ... 
#The class with the highest probability will be the predicted class. Eg : Here ==angry as its probability is 0.9
model.add(Dense(7, activation='softmax'))
```

```python
#compiling
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics='accuracy')
```

```python
#will take a bit of time
model.fit(x=x_train, y=y_train, batch_size=128, epochs=2, validation_data=(x_test, y_test))
```

```python
#saving model
model_json=model.to_json()
with open("emotiondetector.json",'w') as json_file:
    json_file.write(model_json)
model.save("emotiondetector.h5")
```

```python
#Ideally 100 epochs for 74% accuracy but here only 2 epochs as it takes a lot of time
#74% accuracy files provided to use for further real time emotion detection
```

```python
#Checking if model is predicting correct
#Importing model
from keras.models import model_from_json
```

```python
json_file=open("emotiondetector-74%.json","r")
model_json=json_file.read()
json_file.close()
model=model_from_json(model_json)
model.load_weights("emotiondetector-74%.h5")
```

```python
#numerical labels to categorical
label=['angry','disgust','fear','happy','neutral','sad','surprise']
```

```python
#extract feature to check single image
def ef(image):
    img=load_img(image,grayscale=True)
    feature=np.array(img)
    feature=feature.reshape(1,48,48,1)
    return feature/255.0
```

```python
#selecting random image
image='images/train/angry/22.jpg'
img=ef(image)
pred=model.predict(img)
print(pred)
#selecting the label with highest probability
pred_label=label[pred.argmax()]
print("Prediction",pred_label)
```
